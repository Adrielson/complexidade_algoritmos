```python
import pandas as pd
import matplotlib.pyplot as plt

# Ler o arquivo CSV
df = pd.read_csv('sort_performance_log.csv')

# Função para plotar gráficos de tempo e memória para um dado algoritmo
def plot_algorithm_performance(algorithm_name):
    # Filtrar os dados para o algoritmo especificado
    df_algo = df[df['algorithm'] == algorithm_name]
    
    # Plotar gráfico de tempo
    plt.figure(figsize=(10, 5))
    for case in ['melhor', 'pior', 'medio']:
        df_case = df_algo[df_algo['case'] == case]
        plt.plot(df_case['size'], df_case['time'], marker='o', label=f'Caso {case}')
    plt.title(f'Tempo de execução - {algorithm_name}')
    plt.xlabel('Tamanho da entrada')
    plt.ylabel('Tempo (segundos)')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plotar gráfico de memória
    plt.figure(figsize=(10, 5))
    for case in ['melhor', 'pior', 'medio']:
        df_case = df_algo[df_algo['case'] == case]
        plt.plot(df_case['size'], df_case['memory'], marker='o', label=f'Caso {case}')
    plt.title(f'Consumo de memória - {algorithm_name}')
    plt.xlabel('Tamanho da entrada')
    plt.ylabel('Memória (MB)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Lista de algoritmos para plotar
algorithms = df['algorithm'].unique()

# Gerar gráficos para cada algoritmo
for algo in algorithms:
    plot_algorithm_performance(algo)
```

```python
# Função para criar gráficos de tempo
def plot_time(data):
    plt.figure(figsize=(12, 6))
    for algorithm in data['algorithm'].unique():
        subset = data[data['algorithm'] == algorithm]
        plt.plot(subset['size'], subset['time'], label=algorithm)
    
    plt.xlabel('Tamanho da entrada')
    plt.ylabel('Tempo (segundos)')
    plt.title('Tempo de execução por tamanho da entrada')
    plt.legend()
    plt.grid(True)
    plt.show()

# Função para criar gráficos de memória
def plot_memory(data):
    plt.figure(figsize=(12, 6))
    for algorithm in data['algorithm'].unique():
        subset = data[data['algorithm'] == algorithm]
        plt.plot(subset['size'], subset['memory'], label=algorithm)
    
    plt.xlabel('Tamanho da entrada')
    plt.ylabel('Memória (MB)')
    plt.title('Uso de memória por tamanho da entrada')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Criar gráficos
plot_time(data)
```

```python
plot_memory(data)
```

```python
def plot_time_by_case(data, case):
    plt.figure(figsize=(12, 6))
    case_data = data[data['case'] == case]
    for algorithm in case_data['algorithm'].unique():
        subset = case_data[case_data['algorithm'] == algorithm]
        plt.plot(subset['size'], subset['time'], label=algorithm)
    
    plt.xlabel('Tamanho da entrada')
    plt.ylabel('Tempo (segundos)')
    plt.title(f'Tempo de execução ({case}) por tamanho da entrada')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_memory_by_case(data, case):
    plt.figure(figsize=(12, 6))
    case_data = data[data['case'] == case]
    for algorithm in case_data['algorithm'].unique():
        subset = case_data[case_data['algorithm'] == algorithm]
        plt.plot(subset['size'], subset['memory'], label=algorithm)
    
    plt.xlabel('Tamanho da entrada')
    plt.ylabel('Memória (MB)')
    plt.title(f'Uso de memória ({case}) por tamanho da entrada')
    plt.legend()
    plt.grid(True)
    plt.show()

# Criar gráficos para cada caso
cases = ['melhor', 'pior', 'medio']
for case in cases:
    plot_time_by_case(data, case)
    plot_memory_by_case(data, case)
```
