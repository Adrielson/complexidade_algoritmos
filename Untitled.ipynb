```python
import pandas as pd
import matplotlib.pyplot as plt

# Ler o arquivo CSV
df = pd.read_csv('sort_performance_log.csv')

# Funo para plotar grficos de tempo e memria para um dado algoritmo
def plot_algorithm_performance(algorithm_name):
    # Filtrar os dados para o algoritmo especificado
    df_algo = df[df['algorithm'] == algorithm_name]
    
    # Plotar grfico de tempo
    plt.figure(figsize=(10, 5))
    for case in ['melhor', 'pior', 'medio']:
        df_case = df_algo[df_algo['case'] == case]
        plt.plot(df_case['size'], df_case['time'], marker='o', label=f'Caso {case}')
    plt.title(f'Tempo de execuo - {algorithm_name}')
    plt.xlabel('Tamanho da entrada')
    plt.ylabel('Tempo (segundos)')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plotar grfico de memria
    plt.figure(figsize=(10, 5))
    for case in ['melhor', 'pior', 'medio']:
        df_case = df_algo[df_algo['case'] == case]
        plt.plot(df_case['size'], df_case['memory'], marker='o', label=f'Caso {case}')
    plt.title(f'Consumo de memria - {algorithm_name}')
    plt.xlabel('Tamanho da entrada')
    plt.ylabel('Memria (MB)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Lista de algoritmos para plotar
algorithms = df['algorithm'].unique()

# Gerar grficos para cada algoritmo
for algo in algorithms:
    plot_algorithm_performance(algo)
```

```python
# Funo para criar grficos de tempo
def plot_time(data):
    plt.figure(figsize=(12, 6))
    for algorithm in data['algorithm'].unique():
        subset = data[data['algorithm'] == algorithm]
        plt.plot(subset['size'], subset['time'], label=algorithm)
    
    plt.xlabel('Tamanho da entrada')
    plt.ylabel('Tempo (segundos)')
    plt.title('Tempo de execuo por tamanho da entrada')
    plt.legend()
    plt.grid(True)
    plt.show()

# Funo para criar grficos de memria
def plot_memory(data):
    plt.figure(figsize=(12, 6))
    for algorithm in data['algorithm'].unique():
        subset = data[data['algorithm'] == algorithm]
        plt.plot(subset['size'], subset['memory'], label=algorithm)
    
    plt.xlabel('Tamanho da entrada')
    plt.ylabel('Memria (MB)')
    plt.title('Uso de memria por tamanho da entrada')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Criar grficos
plot_time(data)
```

```python
plot_memory(data)
```

```python
def plot_time_by_case(data, case):
    plt.figure(figsize=(12, 6))
    case_data = data[data['case'] == case]
    for algorithm in case_data['algorithm'].unique():
        subset = case_data[case_data['algorithm'] == algorithm]
        plt.plot(subset['size'], subset['time'], label=algorithm)
    
    plt.xlabel('Tamanho da entrada')
    plt.ylabel('Tempo (segundos)')
    plt.title(f'Tempo de execuo ({case}) por tamanho da entrada')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_memory_by_case(data, case):
    plt.figure(figsize=(12, 6))
    case_data = data[data['case'] == case]
    for algorithm in case_data['algorithm'].unique():
        subset = case_data[case_data['algorithm'] == algorithm]
        plt.plot(subset['size'], subset['memory'], label=algorithm)
    
    plt.xlabel('Tamanho da entrada')
    plt.ylabel('Memria (MB)')
    plt.title(f'Uso de memria ({case}) por tamanho da entrada')
    plt.legend()
    plt.grid(True)
    plt.show()

# Criar grficos para cada caso
cases = ['melhor', 'pior', 'medio']
for case in cases:
    plot_time_by_case(data, case)
    plot_memory_by_case(data, case)
```
